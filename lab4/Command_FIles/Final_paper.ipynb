```python
import pandas as pd 
import openpyxl
import numpy as np
print("Data set nr: ",(len("Kardaszewicz")%5)+1)
```

```python
billboard = pd.read_csv('../Original_Data/billboard.csv',encoding= 'unicode_escape',index_col=None)
billboard.dropna(axis=1,how='all')
display((billboard))
```

```python
billboard.rename(columns={'artist.inverted' : 'artist'},inplace=True)
b = billboard.sort_values(by = "artist")
b2 = b.loc[:, ["artist", "track", "time"]]
i = 0
id = []
while i < len(b2):
    id.append(i)
    i += 1
b2.insert(0, 'id', id)
print(b2.to_string(index=False))

```

```python
b.rename(columns={'x1st.week' : 'wk1','x2nd.week' : 'wk2', 'x3rd.week' : 'wk3'},inplace=True)
b3 = b.loc[:, ["year", "artist", "track", "time","date.entered","wk1","wk2","wk3"]]
print(b3.to_string(index=False))
b3.to_csv('../Analysys_Data/first_3weeks.csv')
```
