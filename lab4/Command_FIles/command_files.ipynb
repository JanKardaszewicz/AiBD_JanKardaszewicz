```python
import pandas as pd 
import openpyxl
import numpy as np
print("Data set nr: ",(len("Kardaszewicz")%5)+1, " - billboard")
```

```python
billboard = pd.read_csv('../Original_Data/billboard.csv',encoding= 'unicode_escape',index_col=None)
billboard.rename(columns={'artist.inverted' : 'artist'},inplace=True)
display(billboard)
```

```python
b = billboard.loc[:, ["artist", "track", "time"]]
display(b)

b.to_csv('../Analysys_Data/tracks_on_list.csv')
```

```python
tidy = pd.DataFrame({
    'artist': [billboard.values[0][1]], 'track': [billboard.values[0][2]],
    'time': [billboard.values[0][3]],'date': [billboard.values[0][5]],'week': int(1), 'rank': int(billboard.values[0][7])
})
display(tidy)
```

```python
row, col = billboard.shape
for i in range(row):
    for j in range(7, col):
        if not np.isnan(billboard.loc[i][j]) and (i != 0 or j != 7):
            labels = tidy.shape[0]
            tidy.loc[labels, :] = [billboard.values[i][1], billboard.values[i][2], billboard.values[i][3], billboard.values[i][5], int(j-6), billboard.iloc[i][j]]
display(tidy)
tidy.to_csv('../Analysys_Data/tidy_data_as_lecture_example.csv')
```
