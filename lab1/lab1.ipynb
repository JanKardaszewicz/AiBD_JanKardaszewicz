#### 1. Zdefiniowanie fukcji i wyświetlenie pierwszego przedziału 
   
Do wykonania ćwicznia wykorzystuje biblioteki matplotlib oraz numpy 

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x : x**2 + 5

plt.figure()
x1 = np.linspace(-1, 1)
plt.plot(x1, f(x1))
plt.title("Wykres funkcji w przedziale [-1,1]")
plt.legend("f(x)")
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show()   
```

#### 2. Wyświetlenie drugiego przedziału

```python
plt.figure()
x2 = np.linspace(-6,6)
plt.plot(x2, f(x2))
plt.title("Wykres funkcji w przedziale [-6,6]")
plt.legend("f(x)")
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show() 
```

#### 3. Wyswietlenie trzeciego przedziału

```python
plt.figure()
x3 = np.linspace(0, 5)
plt.plot(x3, f(x3))
plt.title("Wykres funkcji w przedziale [0,5]")
plt.legend("f(x)")
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show() 
```
